## Пропущені значення

```python
from IPython.display import display, HTML
```

```python
# Виведення всіх ознак з пропущеними значеннями

missing_values = df_train_fe.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_percentage = (missing_values / len(df_train_fe)) * 100
missing_df = pd.DataFrame({'Ознака': missing_values.index, 'Кількість відсутніх': missing_values.values, 'Відсоток відсутніх': missing_percentage.values})
missing_df = missing_df.sort_values(by='Кількість відсутніх', ascending=False).reset_index(drop=True)
pd.set_option('display.max_rows', None)
# print(missing_df)
missing_df.to_csv('missing_features_report.csv', index=False)


html_missing_df = missing_df.to_html()
scrolling_div = f"""
<div style="max-height: 200px; overflow-y: scroll; border: 1px solid black; padding: 10px;">
    {html_missing_df}
</div>
"""
display(HTML(scrolling_div))
```

```python
# Вивчення пропущених значень у розрізі класів 

df_class_0 = df_train_fe[df_train_fe['target'] == 0]
df_class_1 = df_train_fe[df_train_fe['target'] == 1]

missing_class_0 = df_class_0.isnull().sum()
missing_class_1 = df_class_1.isnull().sum()

missing_class_0 = missing_class_0[missing_class_0 > 0]
missing_class_1 = missing_class_1[missing_class_1 > 0]

missing_percentage_class_0 = (missing_class_0 / len(df_class_0)) * 100
missing_percentage_class_1 = (missing_class_1 / len(df_class_1)) * 100

missing_df_class_0 = pd.DataFrame({
    'Ознака': missing_class_0.index, 
    'Кількість відсутніх (клас 0)': missing_class_0.values, 
    'Відсоток відсутніх (клас 0)': missing_percentage_class_0.values
})

missing_df_class_1 = pd.DataFrame({
    'Ознака': missing_class_1.index, 
    'Кількість відсутніх (клас 1)': missing_class_1.values, 
    'Відсоток відсутніх (клас 1)': missing_percentage_class_1.values
})


missing_df = pd.merge(missing_df_class_0, missing_df_class_1, on='Ознака', how='outer').fillna(0)

missing_df = missing_df.sort_values(by='Кількість відсутніх (клас 0)', ascending=False).reset_index(drop=True)

pd.set_option('display.max_rows', None)

missing_df.to_csv('missing_features_comparison_report.csv', index=False)

html_missing_df = missing_df.to_html()
scrolling_div = f"""
<div style="max-height: 200px; overflow-y: scroll; border: 1px solid black; padding: 10px;">
    {html_missing_df}
</div>
"""
display(HTML(scrolling_div))
```
